```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('target', axis=1)
x.drop('id', axis=1, inplace=True)
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=16)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo_lr = LogisticRegression(max_iter=8000)
modelo_lr.fit(x_train, y_train)
```

```python
y_prev_lr = modelo_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acr_lr = accuracy_score(y_test, y_prev_lr)
```

```python
print(f'AcurÃ¡cia do modelo com Logistic Regression: {acr_lr}')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_dt = DecisionTreeClassifier()
modelo_dt.fit(x_train, y_train)
```

```python
y_prev_dt = modelo_dt.predict(x_test)
```

```python
acr_dt = accuracy_score(y_test, y_prev_dt)
```

```python
print(f'AcurÃ¡cia do modelo com Decision Tree: {acr_dt}')
```

Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
modelo_svm = SVC()
modelo_svm.fit(x_train, y_train)
```

```python
y_prev_svm = modelo_svm.predict(x_test)
```

```python
acr_svm = accuracy_score(y_test, y_prev_svm)
```

```python
print(f'AcurÃ¡cia do modelo com Support Vector Machine: {acr_svm}')
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
modelo_nb = GaussianNB()
modelo_nb.fit(x_train, y_train)
```

```python
y_prev_nb = modelo_nb.predict(x_test)
```

```python
acr_nb = accuracy_score(y_test, y_prev_nb)
```

```python
print(f'AcurÃ¡cia do modelo com Naive Bayes: {acr_nb}')
```

AcurÃ¡cia Geral

```python
acr = {'Logic Regression': acr_lr, 'Decision Tree': acr_dt, 'Support Vector Machine': acr_svm, 'Naive Bayes': acr_nb}
```

```python
df_acr = pd.DataFrame(acr, ['AcurÃ¡cia'])
```
